```python
def fib_recursive(n):
    if n <=1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

# Time Complexity: O(2^n)
# Space Complexity: O(n) – due to recursion stack

def fib_dp(n):
    if n<=1:
        return n
    dp = [0, 1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

# Time Complexity: O(n)
# Space Complexity: O(n) – uses a list to store all results
```

```python
print("Fibonacci Recursive (n=10):", fib_recursive(10))
print("Fibonacci DP (n=10):", fib_dp(10))
```

```python
def merge_sort(arr):
    if len(arr) <=1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

        result.extend(left[i:])
        result.extend(right[j:])
        return result
    
# Time Complexity: O(n log n)
# Space Complexity: O(n) 
```

```python
arr = [64,25,12,22,11]
print("Merge Sort:", merge_sort(arr.copy()))
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Time Complexity:
#   - Best & Average Case: O(n log n)
#   - Worst Case: O(n^2) 
# # Space Complexity: O(n)
```

```python
arr = [84,28,13,23,12]
print("Quick Sort:", quick_sort(arr.copy()))
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
            arr[j+1] = key
        return arr
        
#Time Complexity:
#Best Case: O(n)
#Worst Case: O(n²)
#Average Case: O(n²)

#Space Complexity: O(1)
```

```python
arr = [77,23,15,16,19]
print("Insertion Sort:", insertion_sort(arr.copy()))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1],arr[j]
        return arr
    
#Time Complexity:
#Best Case: O(n)
#Worst Case: O(n²)
#Average Case: O(n²)
#Space Complexity:O(1)
```

```python
arr = [78,26,16,19,20]
print("Bubble_Sort:", bubble_sort(arr.copy()))
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
               min_idx = j
        arr[i], arr[min_idx] - arr[min_idx], arr[j]
        return arr
    
#Time Complexity:
#Best Case:O(n²)
#Worst Case: O(n²)
#Average Case: O(n²)
#Space Complexity:O(1)
```

```python
arr = [79,27,17,20,21]
print("Selection_Sort:", selection_sort(arr.copy()))
```

```python
def binary_search(arr, target):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

#Time Complexity:O(log n)
#Space Complexity:O(1)
```

```python
arr = [80,28,18,22,24]
print ("Search 18:", binary_search(arr, 18))
```

```python
def measure_time(func, *args):
    start = time.time()
    func(*args)
    end = time.time()
    return end - start

def measure_memory(func, *args):
    mem = memory_usage((func, args), max_iterations=1)
    return max(mem) - min(mem)
```

```python
measure_time(merge_sort, [4,2,3,6,5])
```

```python
measure_memory(merge_sort, [4,2,3,6,5])
```

Algorithm Summary Table
| Algorithm                           | Input                    | Output               | Time Complexity (Best / Avg / Worst)     | Space Complexity | Suitability                        | Trade-offs                                               |
| ----------------------------------- | ------------------------ | -------------------- | ---------------------------------------- | ---------------- | ---------------------------------- | -------------------------------------------------------- |
| **Fibonacci (Recursive)**           | n (integer)              | nth Fibonacci number | – / – / **O(2ⁿ)**                        | **O(n)**         | Educational, small inputs          | Extremely slow for large n due to repeated recomputation |
| **Fibonacci (Dynamic Programming)** | n (integer)              | nth Fibonacci number | O(n) / **O(n)** / O(n)                   | **O(1)**         | Large input calculations           | Uses iterative method instead of simple recursion        |
| **Merge Sort**                      | Array/list               | Sorted list          | **O(n log n)** / O(n log n) / O(n log n) | **O(n)**         | Large datasets, stable sorting     | Requires extra memory                                    |
| **Quick Sort**                      | Array/list               | Sorted list          | O(n log n) / **O(n log n)** / O(n²)      | **O(log n)**     | Fast general sorting               | Worst case bad if pivot chosen poorly                    |
| **Insertion Sort**                  | Array/list               | Sorted list          | **O(n)** / O(n²) / O(n²)                 | **O(1)**         | Nearly sorted or very small inputs | Very slow for large datasets                             |
| **Bubble Sort**                     | Array/list               | Sorted list          | **O(n)** / O(n²) / O(n²)                 | **O(1)**         | Teaching purposes                  | Very inefficient in practice                             |
| **Selection Sort**                  | Array/list               | Sorted list          | O(n²) / O(n²) / **O(n²)**                | **O(1)**         | Low memory scenarios               | Very slow, always quadratic                              |
| **Binary Search**                   | Sorted list + target key | Element index / −1   | **O(1)** / O(log n) / O(log n)           | **O(1)**         | Fast searching on sorted data      | Requires sorted input                                    |
